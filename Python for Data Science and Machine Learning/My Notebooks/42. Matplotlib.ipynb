```python
# First import the plotting library
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set it up to be viewed inline
%matplotlib inline
```

```python
# create a few number sets for plotting

x = np.linspace(1,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# now plot the x vs y
plt.plot(x,y,'r')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('X vs Y')
plt.grid(True,)
#plt.show
```

```python
# We can also do subplots
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.xlabel('X value')
plt.ylabel('Y value')

plt.subplot(1,2,2)
plt.plot(y,x,'r')
plt.xlabel('Y value')
plt.ylabel('X value')
```

```python
# Now lets try the object oriented way fo doing this

# Create a blank canvas
fig = plt.figure()

# Add a set of axes
axes = fig.add_axes([0.1,0.1,1,1]) # axes(left,bottom,width,height)

# Note: The axes are in percent relation of the canvas.

# Plot on the newly created set of axes
axes.plot(x,y,'b')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('OO plot')
```

```python
# now let's try with two sets of axes
fig = plt.figure()

# you can place a plot inside a plot
axes1 = fig.add_axes([0.1,0.1,1,1])
axes2 = fig.add_axes([0.2,0.6,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x,'b')

axes1.set_title('Large graph')
axes2.set_title('Small graph')

axes1.grid(True)
axes2.grid(True)
```

```python
# subplots will automatically arrange the data but can be cramped
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.xlabel('X values')
axes.ylabel('Y values')
```
