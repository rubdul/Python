# K means clustering is an unsupervised learning algorithm that groups similar features together in your data

### can be used in problems such as 
Cluster Similar Documents <br>
Cluster Customers based on Features <br>
Market Segmentation <br>
Identify similar physical groups <br>

```python
# The goal is to divide data into similar groups such that observations in the groups are similar
```

#### process

1. Choose a number of Clusters "K" <br>
2. Randomly assign each point to a cluster <br>
3. Until clusters stop changing, repeat the following <br>

```python
# for each cluster, compute the cluster centroid by taking the mean vector of points in the cluster
# Assign each data point to the cluster for which the centroid is the closest
```

```python
# there is no easy answer for choosing a K value
# you can use the Elbow Method
```

### Elbow method:

Compute the sum of squared error (SSE) for rising values of K etc 2,4,6,8

The SSE is defined as the sum of the squared distance between each member of the cluster and its centroid

Plot K against the SSE and the error should decrease to an asymptote

```python
 # first let's import the usual suspects
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
import matplotlib.axis as Axis
```

```python
# The point of the 
```

```python
from sklearn.datasets import make_blobs
```

```python
# make a different the 
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 101)
```

```python
data [0]
```

```python
data[0].shape
```

```python

plt.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap = 'rainbow')
plt.grid(color = 'blue')
```

```python
# now lets use sci kit learn to create a K means algorithm
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
# now we can get back the labels that the Kmeans algorithm reports for the data
kmeans.labels_
```

```python
# if we didn't know the values we would be done at this point.
```

```python
# let's compare our K means data to the Kmeans label data labels

fig, (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_,cmap = 'rainbow')
ax1.grid()

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'rainbow')
ax2.grid()
```
