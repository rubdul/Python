```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
#create a bar plot of the mean bill amount based on sex
sns.barplot(x='sex',y='total_bill', data=tips)
```

```python
# Change the graph to consider standard deviation as the measurement
sns.barplot(x='sex',y='total_bill', data=tips,estimator = np.std)
```

```python
# Now let's do a countplot which is similar to a bar plot but counts the number of occurrences
sns.countplot(x='sex',data=tips)
```

```python
# Box plots and violin plots show the distribution of quantitative data in a series
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# violin plots show the kernel density estimation of the underlying distribution
# It's as if the data was read on it's side
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python
# Now let's draw a one varible categorical scatter plot aka a Strip Plot
sns.stripplot(x='day', y='total_bill',data=tips)
```

```python
# use the jitter parameter to make this easier to read by inputting noise
sns.stripplot(x='day', y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
# A swarm plot is like a strip plot but the points are adjusted to prevent overlap
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
# You can also combine violin and swarm plots 
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips,color='black')
```

```python
# Now elts do a factor plot
sns.factorplot()
```
