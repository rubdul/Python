```python
# Ln plot function allows us to display linear models using seaborn
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y = 'tip', data=tips, hue='sex')
```

```python
# to change the size of the markers you can import a dictionary
sns.lmplot(x='total_bill',y = 'tip', data=tips, hue='sex', markers=['o','v'], scatter_kws = {'s':100})
```

```python
# you can also separate the plot by columns and rows
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex', row = 'time', hue = 'sex')
```

```python
# change the aspect and size ratio
sns.lmplot(x='total_bill',y='tip',data=tips,col='day', hue = 'sex', aspect =0.6, size =10)
```
