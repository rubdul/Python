```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# you can use the set_style method to modify the style of the graph prior to entry
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
# use the despine operator to reduce the size argument
# by default it will remove the top and right ones.
sns.despine()
```

```python
# IF you want to change the aspect ratio and size of the plot use the figsize parameter
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
#you can also use the set context method to alter size
sns.set_context('notebook')
sns.countplot(x='sex',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip', data = tips, hue='sex',palette='coolwarm')
```
