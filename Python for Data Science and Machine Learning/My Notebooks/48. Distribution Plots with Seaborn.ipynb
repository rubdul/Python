```python
# first import seaborn and setup matplotlib as inline
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# use the distplot method to create a distribution plot based on a single column and get a kernal distribution estimate
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
# We can also use jointplot for bivariate data
# You can change the kind of graph to 'hex', 'reg', or even 'kde' 
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# Regression plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
# You can make a pair plot for different conditions and multiple attributes of the data
# It's a quick way for you to visuallize your data
sns.pairplot(tips)
```

```python
# You can use the hue parameter to make colors based on a specific attribute
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
# Now let's do rug plots
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```
