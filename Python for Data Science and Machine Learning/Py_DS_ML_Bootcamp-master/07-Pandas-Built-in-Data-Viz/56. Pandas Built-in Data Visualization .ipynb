```python
# pandas has some built in data plot like histograms
df1['A'].hist(bins=30)
```

```python
# If you import Seaborn first, your plots will look different
df1['A'].hist(bins=30)
```

```python
df1['A'].plot.hist(bins=50)
```

```python
# Area plot
df2.plot.area(alpha=0.4)
```

```python
# bar plot
df2.plot.bar(stacked=True)
```

```python
# line plot
# Remember now a days not to specify the x= parameter or it will cause an error 
df1.plot.line(y='B',figsize=(12,3),lw=1)
```
