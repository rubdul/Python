## Task Two: Download the PDF from the Google Drive link and find the phone number that is in the document. 

```python
# You should get this phone number
# 505 503 4455
```

```python
# now look into the PDF and extract it's page text and use regex to look for the phone number
```

```python
pdfpath =  root_name + '\\Find_the_Phone_Number.pdf'
```

```python
file = open(pdfpath, 'rb')
```

```python
pdf = PyPDF2.PdfReader(file)
```

```python
# extract the text in a for loop
pdf.numPages
```

```python
text = ""

for pagenum in range(pdf.numPages):
    text = text + str(pdf.getPage(pagenum).extract_text())

words = print(text)

```

```python
text
```

```python
# now find the phone number

phone = re.search(r'\d{3}.\d{3}.\d{4}',text)
print(phone)
```
